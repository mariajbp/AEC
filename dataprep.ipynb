# Data Analysis and Visualiation

### Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

### Load Train Dataset

```python
salaryClassification = pd.read_csv('data/training.csv', sep = ';')

salaryClassification = salaryClassification.replace(' ?', np.NaN)
```

### Peak Dataset

```python
salaryClassification.head(50)
```

### Descriptive Statistics

```python
salaryClassification.info()
```

```python
salaryClassification.describe(include='all')
```

### Check Missing Values

```python
salaryClassification.isna().sum()
```

### Treat Missing Values

Tratamento das features workclass, occupation e native-country

```python
salaryClassification[' workclass'] = salaryClassification[' workclass'].replace(np.NaN, 'Unknown')
```

```python
salaryClassification[' occupation'] = salaryClassification[' occupation'].replace(np.NaN, 'Other')
```

```python
salaryClassification[' native-country'] = salaryClassification[' native-country'].replace(np.NaN, 'Other')
```

```python
salaryClassification.head()
```

# EDA

```python
sns.set_style('whitegrid')
ax = sns.countplot(x=' salary-classification', data = salaryClassification, palette=['#d8bce8',"#FAAE7B"])

total = len(salaryClassification[' salary-classification'])

for p in ax.patches:
    height = p.get_height()
    ax.text(p.get_x()+p.get_width()/2.,height + 3,'{:.1f}%'.format(100 * height/total),ha="center")
```

### Outliers Analysis

### Correlação das carateristicas numéricas

```python
fig, ax = plt.subplots(figsize=(25, 25))
p = sns.heatmap(salaryClassification.corr(), annot=True, cmap="BuPu")
plt.title("Correlation of Numerical Features", fontsize=20)
plt.show()
```

### Classificação dos salários em função do número de anos de educação

```python
fig, ax = plt.subplots(figsize=(10, 10))
sns.countplot(x = ' salary-classification', hue = ' education-num', data = salaryClassification)
```

### Classificação dos salários em função do grau de educação

```python
fig, ax = plt.subplots(figsize=(10, 10))
sns.countplot(x = ' salary-classification', hue = ' education', data = salaryClassification)
```

### Classificação dos salários em função da idade

```python
salaryClassification['age'] = pd.cut(salaryClassification['age'], bins = [0, 25, 50, 100], labels = ['Young', 'Adult', 'Old'])
sns.countplot(x = ' salary-classification', hue = 'age', data = salaryClassification, palette=['#d8bce8',"#FAAE7B", "#97c6e8"] )
```

### Classificação dos salários em função da profissão

```python
fig, ax = plt.subplots(figsize=(10, 10))
sns.countplot(x = ' salary-classification', hue = ' occupation', data = salaryClassification)
```

### Classificação dos salários em função da diferença de capital

```python
salaryClassification['capital-diff'] = salaryClassification[' capital-gain'] - salaryClassification[' capital-loss']
salaryClassification.drop([' capital-gain'], axis = 1, inplace = True)
salaryClassification.drop([' capital-loss'], axis = 1, inplace = True)
```

```python
salaryClassification['capital-diff'] = pd.cut(salaryClassification['capital-diff'], bins = [-5000, 5000, 100000], labels = ['Minor', 'Major'])
sns.countplot(x = ' salary-classification', hue = 'capital-diff', data = salaryClassification, palette=['#d8bce8', "#97c6e8"])
```

### Classificação dos salários em do número de horas de trabalho semanais

```python
salaryClassification[' hours-per-week'] = pd.cut(salaryClassification[' hours-per-week'], bins = [0, 30, 40, 100], labels = ['Lesser Hours', 'Normal Hours', 'Extra Hours'])

sns.countplot(x = ' salary-classification', hue = ' hours-per-week', data = salaryClassification, palette=['#d8bce8',"#97e8bb", "#97c6e8"])
```
